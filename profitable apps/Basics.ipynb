## Opening Data and exploring data

```python
#first, let's make function to open the dataset

def open_data(data, header = True):
    #read_in_data
    opendata = open(data)
    from csv import reader
    read_data = reader(opendata)
    
    #transform data into list of lists
    new_data = list(read_data)
    if header:
        return new_data[0], new_data[1:]
    
    return new_data
```

```python
#open apple dataset
applestore_data_header, applestore_data = open_data('AppleStore.csv')

#open google dataset
googleplay_data_header, googleplay_data = open_data('googleplaystore.csv')
```

we can print many rows dataset using function below

```python
#let's make function for explore dataset
def explore_data(dataset, start, end, rows_and_columns=False):
    dataset_slice = dataset[start:end]
    
    #print row in dataset_slice
    for row in dataset_slice:
        print(row)
        print('\n')
        
    #print number of row if rows_and columns true
    if rows_and_columns:
        print('number of rows: ', len(dataset))
        print('number of columns: ', len(dataset[0]))
```

```python
#function to find duplicate data
def find_duplicate(data):
    duplicate_app =[]
    unique_app =[]
    
    #if name not in duplicate_app , input name into unique_app, if in name in duplicate data input name into dupicate app
    for app in data:
        name = app[0]
        if name in unique_app:
            duplicate_app.append(name)
        else:
            unique_app.append(name)
            
    return duplicate_app, unique_app
            
```

To remove the duplicates, we will:

Create a dictionary, where each dictionary key is a unique app name and the corresponding dictionary value is the highest number of reviews of that app.

Use the information stored in the dictionary and create a new data set, which will have only one entry per app (and for each app, we'll only select the entry with the highest number of reviews).

```python
#create dictionary
reviews_max = {}

#loop the row in dataset
for app in googleplay_data:
    name = app[0]
    n_reviews = float(app[3])
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
    elif name not in reviews_max :
        reviews_max[name] = n_reviews
        
print(len(reviews_max))
        
        
```

```python
#clean data from duplicate data
googleplay_clean =[]
already_added = []

for apps in googleplay_data:
    name = apps[0]
    n_reviews = float(apps[3])
    
    if (reviews_max[name] == n_reviews) and (name not in already_added):
        googleplay_clean.append(apps)
        already_added.append(name)
    
print(len(googleplay_clean))
```

To minimize the impact of data loss, we'll only remove an app if its name has more than three characters with corresponding numbers falling outside the ASCII range. 

```python

def is_english(a_string):
    non_ascii = 0
    for char in a_string:
        if ord(char) > 127:
            non_ascii += 1
    
    if non_ascii > 3:
        return False
    else:
        return True
    
print(is_english('Instagram'))
print(is_english('Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠'))
print(is_english('Docs To Go‚Ñ¢ Free Office Suite'))

english = is_english('Instachat üòú')
print(english)

    
```

# Most Common apps by Genre

 to know , what kind of apps that will be more attracted by people , i focus on prime_genre column in the app store dataset and genres, category columns of the googleplay dataset

we will build two function
1. one function to generate frequency tables that show percentages
2. other function we can use to display the percentages in a descending order

```python
#make frequency table
def freq_table(dataset, index):
    freq_dictionary = {}
    total = 0
    for app in dataset:
        total += 1
        data = app[index]
        if data in freq_dictionary:
            freq_dictionary[data] += 1
        else:
            freq_dictionary[data] = 1
            
    table_percentages = {}
    for key in freq_dictionary:
        percentages = (freq_dictionary[key] / total) * 100
        percent_round = round(percentages, 4)
        table_percentages[key] = percent_round
    
    return table_percentages
```

```python
#def display dataset
import pandas as pd
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display =[]
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)
        
    #sorted the data from the highest
    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ': ', entry[0])
```

```python

import numpy as np

#make new_list to distict it with source data
no_nan =[]
for app in googleplay_final:
    no_nan.append(app)
    
#replace nan with 0
for app in no_nan:
    if app[2] == 'NaN':
        app[2] = '0'

display_table(no_nan,2)
```
