### compare the number of men with the number of women in each category of majors.

to begin with, let's make many function

```python
#make dictionary to split the data
category_men ={}
category_women ={}
n_men =0
n_women =0

#using unique to delete duplicate data
major_category = recent_grads['Major_category'].unique()

#loop the every major, in every major sum number of men and women and put it in dictionary
for major in major_category:
    new_data = recent_grads[recent_grads['Major_category']==major]
    n_men = new_data['Men'].sum()
    n_women = new_data['Women'].sum()
    category_men[major] = n_men
    category_women[major] = n_women
```

```python
#make new series using new dictionary
major_men_series = pd.Series(category_men)
major_women_series = pd.Series(category_women)

#concat the both of series , become dataframe
men_and_women_major = pd.DataFrame({'total_men': major_men_series, 'total_women': major_women_series})
men_and_women_major
```

```python

#make bar plot
ax = men_and_women_major.plot.bar(figsize=(15,10),rot=90)

```

from the data we know that majority women and men have diferent pattern, biggest total women and women has happen in bussines category when smallest total in interdisplinary category.

we also know, that 10 of major_category have total women than men and only six category have more men than women.

many of bar plot have big different too, for example , in engineering category, total of men have big different with total women, when education, social work and health have more women , rather than men, with big different, 

for conclution, may be this pattern can happen caused by jobdesk of major category itself and also habit which developing in society, 
for instance, many engineering's job, have outdoor jobdesk, so it will be more men than women
